#**Titanic Data Analysis**#

**STEP 1: Load Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Vaibhav Gavade\\Downloads\\test_data.csv")
print(df.head())
```

```python
# Fix Gender (0/1 → Male/Female)
df['Sex'] = df['Sex'].map({0: 'Female', 1: 'Male'})
```

```python
# Fix Passenger Class (one-hot → single column)
df['Pclass'] = df[['Pclass_1','Pclass_2','Pclass_3']].idxmax(axis=1)
df['Pclass'] = df['Pclass'].str[-1] 
```

**STEP 2:** Analysis Questions

Aim:To analyze the survival patterns of passengers based on gender (male vs female).

```python
print("\nSurvival by Gender:")
print(df.groupby('Sex')['Survived'].mean())
```

**Result:** VisualizationsFemale passengers had a significantly higher survival rate compared to male passengers.

**STEP 3:** Visualizations

**Aim:** To analyze the survival pattern of passengers based on gender (male and female) using the Titanic Dataset.

```python
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

**Result:** VisualizationsFrom the graph, it is observed that female passengers had a significantly higher survival rate compared to male passengers. This indicates that gender played an important role in survival, with females being more likely to survive than males.

**Aim**To study survival based on passenger class.

```python
print("\nSurvival by Class:")
print(df.groupby('Pclass')['Survived'].mean())

sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Class")
plt.show()
```

**Result**From the graph, it is observed that passengers in 1st class had the highest survival rate, followed by 2nd class, while 3rd class passengers had the lowest survival rate. This indicates that passenger class significantly affected survival chances, with higher-class passengers being more likely to survive.

**Aim**To study the distribution of passenger ages using the Titanic Dataset through a histogram.

```python
plt.hist(df["Age"])
plt.title("Passenger Ages (Scaled)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

**Result**From the histogram, it is observed that most passengers fall within the middle age range, with fewer passengers at very low and very high age values. Since the age data is scaled, the exact age values cannot be interpreted directly, but the overall distribution shows that the majority of passengers were adults.
